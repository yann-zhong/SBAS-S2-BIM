<b>Exercice 2</b>: Score alignement global, local <br>
Faire une fonction pour calculer : 1) la pourcentage d'identité, 2) la pourcentage de similarité, 3) le score d’un alignement, en utilisant la matrice blosum62. Le calcul de la similarité doit prendre en considération les acides aminés ayant les mêmes propriétés physico-chimiques. Dans notre cas, tous les acides amine qui ont des valeurs supérieur à zéro dans la matrice BLOSUM62. Le score basé sur la matrice blosum doit aussi prendre en considération les deux pénalités de gaps : ouverture et extension. Testez vos fonctions en utilisant le fichier test.fasta


```python
import numpy as np

#variable initiation
aa = ['A', 'R', 'N', 'D', 'C', 'Q', 'E', 'G', 'H', 'I', 'L', 'K', 'M', 'F', 'P', 'S', 'T', 'W', 'Y', 'V', 'B', 'Z', 'X', '-']
q = len(aa)

#parameters
gap_open = -5
gap_ext = -0.5

#files
input_test_f = 'test.fasta'
input_blosum_f = 'BLOSUM62.txt'
```

```python
def read_fasta (input_f):
    
    list_seqs=[]
    fichier=open(input_f,'r')
    for i in fichier:
        if i[0]!=">":
            list_seqs.append(i[:-1])  # don't take the \n line
    
    return list_seqs
    
testAln = read_fasta(input_test_f)
print (testAln)
```

```python
#read Blosum
def read_blosum (input_f, q):
    blosum=[]
    fichier=open(input_f,'r')
    for i in fichier:
        if i[0]!=" " and i[0]!="#":
            blosum.append(i.split()[1:])
    return blosum

matr_62 = read_blosum (input_blosum_f, q)
print(matr_62)
```

```python
# NON OPTIMIZED VERSION
def id_sim_sc(seq1, seq2, gap_open, gap_ext, matr_62, alphabet):
    identity,similarity,score=0,0,0
    
    for ind in range(len(seq1)):
        if seq1[ind] == seq2[ind]:                                # identity: match
            identity += 1
            
        i = seq1[ind]
        j = seq2[ind]
        sc = int(matr_62[alphabet.index(i)][alphabet.index(j)])   # score is calculated as according to BLOSUM matrix
        score += sc
        
        if sc > 0:                                                # similarity is defined as any score greater than 0 
            similarity += 1                                       # representing a non penalized mutation
    identity,similarity = identity/len(seq1),similarity/len(seq1)
    
    return identity, similarity, score

identity, similarity, score = id_sim_sc(testAln[0], testAln[1], gap_open, gap_ext, matr_62, aa)
print ("identity = ", identity , " similarity = ", similarity, " score = ", score )
```

```python
# OPTIMIZED VERSION, USE THIS ONE
def id_sim_sc(seq1, seq2, gap_open, gap_ext, matr_62, alphabet):
    identity,similarity,score=0,0,0
    for ind in range(len(seq1)):
        print(score)
        if seq1[ind]=='-':
            if seq1[ind-1]=='-' and ind>0:
                score+=gap_ext
            if seq1[ind-1]!='-' and ind>0:
                score+=gap_open
        if seq2[ind]=='-':
            if seq2[ind-1]=='-' and ind>0:
                score+=gap_ext
            if seq2[ind-1]!='-' and ind>0:
                score+=gap_open

        if seq1[ind]==seq2[ind] and seq1[ind]!='-':
            identity+=1
            i=seq1[ind]
            sc=int(matr_62[alphabet.index(i)][alphabet.index(i)]) # score is calculated as according to BLOSUM matrix
            score+=sc
            if sc>0:
                similarity+=1                                     # similarity is defined as any score greater than 0 
            continue                                              # representing a non penalized mutation
        if seq1[ind]!='-' and seq2[ind]!='-':
            i,j=seq1[ind],seq2[ind]
            sc=int(matr_62[alphabet.index(i)][alphabet.index(j)])
            score+=sc
            if sc>0:
                similarity+=1
            continue

    identity,similarity=identity/len(seq1),similarity/len(seq1)

    return identity, similarity, score
```

Faire un script (plusieurs fonctions seront nécessaires) pour produire une matrice de substitution à la Blosum. Utiliser l’alignement du fichier <b>blocks.dat</b>.

```python
from itertools import combinations, combinations_with_replacement
from collections import Counter

input_block_f = 'block.dat'

#Alphabet
aa = ['A', 'R', 'N', 'D', 'C', 'Q', 'E', 'G', 'H', 'I', 'L', 'K', 'M', 'F', 'P', 'S', 'T', 'W', 'Y', 'V', 'X']
aa = sorted(aa)
print(aa)

#Generate the dictionary of all possible pairwise combinations (occurrences set to zero for now)
def empty_occ(alphabet):
    occ_dict={}
    occ=list(combinations_with_replacement(alphabet, r=2))
    for i in occ:
        occ_dict[min(i)+max(i)]=0
    return occ_dict  #Dictionary. Keys: pairs of aa types, Values: zero

pairs_occ_dict = empty_occ(aa)
print(pairs_occ_dict)
```

```python
#read alignment file
from io import StringIO 
import sys

""" 'Lazy version'
def readAlnFile(input_f):
    
    aln = []
    fichier = open(input_f,'r')
    for i in fichier:
        aln.append(i[:-1])
    return aln
"""

# Cleaner version
def readAlnFile(input_f):
    aln = []
    file = open(input_f,'r')
    for line in file:
        splitline=[]
        for j in range(len(line)):
            if line[j]!='\n':
                splitline.append(line[j])
        aln.append(splitline)
    file.close
    aln=np.array(aln)
    return aln

aln  = readAlnFile(input_block_f)
print(aln)
```

```python
#compute occ_ij occurrences
import itertools
def occurrences(aln, alphabet):
    occ=empty_occ(alphabet)
    a=list(itertools.permutations(range(len(aln)), r=2))
    for i in range(len(aln[0])):
        col=np.array(aln[:,i])
        for j in a:
            couple=[col[j[0]],col[j[1]]]
            couple.sort()
            occ[couple[0]+couple[1]]+=1
    
    return occ    # Dict with the same keys of empty_occ 

pairs_occ_dict = occurrences(aln, aa)
print(pairs_occ_dict)
```

```python
import itertools

a=list(itertools.permutations(range(3), r=2))
print(a)

# Example of using permutations to check every unique occurence used in the function above
```

```python
#compute total sum of occurrences
def total(occ):
    
    tot=sum(list(occ.values()))
    
    return tot    # int: total sum of occurrences
```

```python
#compute p(i): probability of finding a residue type
def compute_p(pairs_occ_dict, alphabet):
    tot=total(pairs_occ_dict)
    p={}
    for i in pairs_occ_dict:
        p[i]=pairs_occ_dict[i]/tot
    return p    # vector of length len(alphabet)
    
p = compute_p(pairs_occ_dict, aa)
print(p)
```

```python
#compute S(ij)
def compute_S(pairs_occ_dict, alphabet, p):
    S=np.zeros((len(alphabet),len(alphabet)))
    q=[]
    for i in alphabet:
        som=0
        for k in range(len(alphabet)):
            if i!=alphabet[k]:
                couple=[i,alphabet[k]]
                couple.sort()
                som+=p[min(couple)+max(couple)]
        q.append(p[i+i]+som)
    for i in range(len(alphabet)):
        for j in range(len(alphabet)):
            S[i][j]=np.log(p[alphabet[min(i,j)]+alphabet[max(i,j)]]/(2*q[i]*q[j])) # log2 ratio computation

    return S

S = compute_S(pairs_occ_dict, aa, p)
print(S)
```

```python
#Print the matrix in an easily readable format

for i in range(len(S)):
    for j in range(len(S[i])):
        if S[i][j]<0:
            S[i][j]=-1
        else:
            S[i][j]=0
print(S)
```
